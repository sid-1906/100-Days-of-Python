<h3>Using get_dummies</h3>

```python
df1=pd.get_dummies(df["Purchased"])
```

```python
df1
```

```python
df1.columns
```

```python
df2 = df1.set_axis(['Purchased-No', 'Purchased-Yes'], axis=1, inplace=False)
```

```python
df2
```

```python
merged = pd.concat([df, df2], axis='columns')
```

```python
merged
```

```python
merged.drop(['Purchased'], axis='columns')
```
